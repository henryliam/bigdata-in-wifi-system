```python
df_1_train=df_1[['lat','long','alt','signal_strength','bssid']]
```

```python
# lats_uni2=df_1_train2.lat.unique()
lats_uni=df_1_train2['lat']
len(lats_uni)
```

```python
list1=[]

for lat in lats_uni:
    df_mean=df_1_train.loc[df_1_train['lat']==lat]
    list1.append(df_mean.loc[:,'signal_strength'].mean())
```

```python
len(list1)
```

```python
df_1_train2=df_1_train2.drop(['signal_strength'],axis=1)
#df1=df_1_train2.drop_duplicates(["lat"]).reset_index(drop=True)
```

```python
df_1_train2['signal_strength']=list1
```

```python
df_1_train2.round({"lat":6, "long":6,"signal_strength":2})
```

```python

train, test = train_test_split(df_1_train2, test_size=0.1)
df_1_train2.to_csv('RSS.csv')
```

```python
df_X_train = train[['lat','long','alt']]
df_X_test =test[['lat','long','alt']]
df_Y_train = train[['signal_strength']]
df_Y_test = test[['signal_strength']]
```

```python
# print(df_X_train)
# print(df_Y_train)
# print(df_Y_test)
# print(df_X_test)
```

```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(df_X_train,df_Y_train)
```

```python
RSS_y_pred = regr.predict(df_X_test)
```

```python
print("Mean squared error: %.2f"
      % mean_squared_error(df_Y_test, RSS_y_pred))
print('Variance score: %.2f' % r2_score(df_Y_test, RSS_y_pred))
```
